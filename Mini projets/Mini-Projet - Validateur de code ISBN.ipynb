!!! abstract Introduction :

L’ISBN est un numéro international normalisé permettant l’identiﬁcation d’un livre dans une édition donnée.
ISBN est l’acronyme d’International Standard Book Number
L’ISBN a été conçu pour simpliﬁer le traitement informatisé des livres : les libraires peuvent passer des commandes standardisées, les distributeurs ont le même code pour traiter les commandes et les retours, les diﬀérentes opérations de gestion dans les bibliothèques et centres de documentation sont également facilitées. Par ailleurs, le caractère international de cette numérotation constitue, à l’étranger également, une référence unique pour tous les professionnels du livre.
L’ISBN identiﬁe donc de façon unique un livre quel que soit son support de publication,
imprimé ou numérique.

L’ISBN est un numéro à 13 chiﬀres depuis le  1er janvier 2007
(les ouvrages antérieurs avaient un ISBN à 10 chiﬀres).


Les parties de l’ISBN : 978 2 2772 3892 8
* 978 signiﬁe que l’article est un livre (’Bookland’) ( 979 sera aussi employé en fonction des besoins...) Dans un code barre, les 3 premiers chiﬀres indiquent le pays producteur, comme 978 et 979 ne sont pas utilisés, on les a attribués aux livres…
* La deuxième partie de l’ISBN est le numéro identiﬁant le groupe national, linguistique ou géographique. Le 2 identiﬁe la francophonie dans son ensemble.
* 2772 : C’est un numéro qui identiﬁe un éditeur en particulier.
* 3892 : C’est le numéro d’identiﬁcation d’une édition donnée d’une publication produite par un éditeur en particulier.
* 8 : La cinquième partie de l’ISBN est la clé de contrôle. Il s’agit de la partie ﬁnale de l’ISBN qui est reliée aux précédents caractères de la chaîne par un algorithme mathématique et qui permet d’en vériﬁer la validité. La clé peut varier de 0 à 9.

!!! question Première étape : La saisie de l'utilisateur
Proposer une fonction qui va demander à l'utilisateur de saisr un code ISBN de 13 chiffres. Cette fonction vérifiera que la saisie est correcte :

!!! question Deuxième étape : La somme pondérée
Notre code ISBN est une chaîne de 13 caractères : code_isbn = "9782277238928" 

* Il faut faire la somme (s1) des caractères (transformés en nombre) d’indices (0, 2, 4, 6, 8, 10).
* Il faut faire la somme (s2) pondérée par 3 des caractères (transformés en nombre) d’indices (1, 3, 5, 7, 9, 11).
* Il faudra ajouter s1 et s2.

**Ecrire une fonction qui prend en paramètre un code ISBN sous forme d'une chaine de caractère et qui renvoie la somme s1 + s2 :**
!!!

```python
mon_isbn = "9782277238928"
#votre fonction ici 
def somme_ponderee(isbn : str) -> int:
    """Calculer la somme pondérée du code ISBN"""
    s1 = 0
    s2 = 0
    for i in range(1, 13, 2):
        s1 = s1 + int(isbn[i - 1])
    for j in range(2, 13, 2):
        s2 = s2 + (int(isbn[j - 1]) * 3)
    return s1 + s2
```

!!! question Troisième étape : Le calcul de la clé de contrôle
La clé de contrôle se calcule comme suit:
* On détermine le reste de la division de s = s1 + s2 par 10.
* Si ce reste est nul alors la clé est égale à 0.
* Sinon la clé est égale à 10 − reste.    
**Ecrire une fonction qui prend en paramètre un code ISBN sous forme d'une chaine de caractère et qui renvoie la clé de contrôle :**
!!!

```python
mon_isbn = "9782277238928"

def calcul_cle(isbn : str) -> int :
    """Calculer la clé de contrôle du code ISBN"""
    somme = somme_ponderee(isbn)
    reste = somme % 10
    if (reste == 0):
        return 0
    else:
        return 10 - reste
```

!!! question Pour finir : Valider le code ISBN

Il ne reste plus qu’à comparer cette clé calculée avec le 13ème chiﬀre du code ISBN, s’il y a égalité alors le code est valide, sinon il n’est pas valide.
