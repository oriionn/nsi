Dans un jeu de stratégie au tour par tour, les joueurs dirigent des équipent de guerriers qui vont se combattre. Un guerrier est connu par son nom et est caractérisé par :

- **ptsVie :** ses points de vie, qu’il perd au fur et à mesure des combats ;


- **force :** sa force, qui détermine la puissance des coups qu’il prote à l’adversaire ;

- **adresse :** son adresse, qui détermine son habileté à atteindre son adversaire (valeur de 0 à 100) ;

- **nbCoups :** le nombre de coups qu’il peut porter lors d’une attaque ;

- **armure :** son armure, qui le protège et atténue les coups qu’il reçoit, (valeur de 0 à 100).

Dans la suite de l’exercice, nous modéliserons un guerrier en python grâce à un dictionnaire dont les clés prendront les valeurs : nom, ptsVie, force, adresse, nbCoups et armure.



![Guerriers](https://capytale2.ac-paris.fr/web/sites/default/files/2023/01-07/14-59-32/guerriers.jpg)

```python
helmut={'nom':'helmut','ptsVie':100,'force':6,'adresse':75,'nbCoups':3,'armure':30}
olga={'nom':'olga','ptsVie':98,'force':8,'adresse':90,'nbCoups':2,'armure':35}
irina={'nom':'irina','ptsVie':100,'force':2,'adresse':55,'nbCoups':2,'armure':20}
boris={'nom':'boris','ptsVie':100,'force':3,'adresse':30,'nbCoups':3,'armure':15}
```

!!! question Exercice 1. 
Proposez une fonction `est_vivant()` qui teste si un guerrier donné en paramètre est vivant ou non et qui renvoie un booléen.
Un guerrier est vivant tant qu'il a des points de vie.
!!!


```python
def est_vivant(guerrier):
    if 'ptsVie' in guerrier:
        if guerrier['ptsVie'] > 0:
            return True
    return false
```

!!! question Exercice 3.
En réalité, lorsqu’un guerrier donne un coup à un adversaire, la valeur du coup est celle de sa force si l’adversaire est atteint.
Cette possibilité est tirée au hasard, grâce à la probabilité de réussite donnée par l’adresse du guerrier (en % de chance de réussir)

Proposez la fonction `donne_un_coup2()` qui renvoie la valeur du coup porté par un guerrier donné prenant en compte l’adresse du guerrier.
!!!

```python
from random import randint
def donne_un_coup2(guerrier):
    random = randint(0,100)
    if random >= guerrier['adresse']:
        return 0
    else:
        return guerrier['force']
```

```python
donne_un_coup2(boris)
```

!!! question Exercice 4. 
Lorsqu’un guerrier reçoit un coup, son armure le protège. La valeur de son armure est appliquée comme un pourcentage de protection sur la force du coup reçu.
Ainsi, un guerrier qui reçoit un coup de 5 points, et qui a une armure dont le coefficient de protection est de 20%, voit ses points de vie diminuer de seulement 4. $$5−5×\frac{20}{100}=4$$
Proposer la fonction `prend_un_coup()` qui, pour un guerrier et une force de coup donné, diminue les points de vie du guerrier.
!!!

```python
def prend_un_coup(guerrier,force):
    protection = guerrier['armure']
    perdu = force - (force * (protection/100))
    guerrier['ptsVie'] -= perdu
```

```python
prend_un_coup(irina,8)
irina
```

!!! question Exercice 5.Les points de vie d'un guerrier sont des entiers.

Proposer la fonction `prend_un_coup2()`  qui, pour un guerrier et une force de coup donné, diminue les points de vie du guerrier.
!!!

```python
def prend_un_coup2(guerrier,force):
    protection = guerrier['armure']
    perdu = force - (force * (protection/100))
    guerrier['ptsVie'] -= int(perdu)
```

!!! question Exercice 6.
Un guerrier ne peut jamais avoir de points de vie négatifs.   
Proposer la fonction `prend_un_coup3()`  qui, pour un guerrier et une force de coup donné, diminue les points de vie du guerrier et renvoie s’il est mort dès qu’il n’a plus de points de vie.
!!!

```python
def prend_un_coup3(guerrier,force):
    protection = guerrier['armure']
    perdu = force - (force * (protection/100))
    guerrier['ptsVie'] -= int(perdu)
    if guerrier['ptsVie'] < 0:
        return True
    else:
        return False
```
