!!! abstract On peut facilement parcourir un tableau à l'aide d'une boucle <code>for</code>. <br>
Il existe pour cela deux méthodes en Python :<br>
<ul>
    <li>
        Méthode 1 : La variable de boucle parcourt successivement tous les <strong>indices</strong> du tableau.
    </li>
    <li>
        Méthode 2 : La variable de boucle parcourt successivement toutes les <strong>valeurs</strong> du tableau.
    </li>
</ul>
!!!

````python
notes = [12,17,15,14]

# Méthode 1
print("Méthode 1 :")

for i in range(len(notes)) :   # i prend successivement toutes les valeurs des indices du tableau
    print(notes[i])

#Méthode 2
print("Méthode 2 : ")

for n in notes :    # n prend successivement toutes les valeurs du tableau
    print(n)
````

!!! abstract Remarque :
La méthode 2 est préférable car sa syntaxe est plus courte et plus lisible. Cependant, elle ne peut fonctionner que si l'on n'utilise pas l'indice de la valeur considérée dans la boucle. Si par exemple on cherche l'indice correspondant à la plus grande valeur du tableau, seule la méthode 1 pourra être mise en oeuvre.
!!!

<strong>Arrêt prématuré du parcours</strong><br>
Pour certains algorithmes, il est inutile de parcourir entièrement le tableau du début à la fin.
Dans ce cas, il est préférable d'utiliser une boucle <code>while</code> (tant que), qui permet de s'arrêter avant d'avoir atteint la fin du tableau. On économise alors du temps de calcul.<br>
<strong><br>Exemple : </strong> Le programme ci-dessous renvoie <code>True</code> si le tableau contient deux éléments consécutifs égaux, et <code>False</code> sinon :

````python
tab = [2,4,8,7,7,6,4,8,3]      # Supprimez un des deux 7 pour modifier le résultat du programme
i = 1
doublon = False     # ce booléen prendra la valeur True si un doublon est trouvé
while i < len(tab) and not doublon: # on arrête le parcours du tableau si on a déjà trouvé un doublon
    print(i)
    if tab[i] == tab[i-1]:
        doublon = True
    i = i + 1
print(doublon)
````

!!! danger Exercice bilan : Entrelacer deux tableaux
Ecrire un programme qui entrelace les deux tableaux <code>tab1</code> et <code>tab2</code> ci-dessous, et affiche le tableau entrelacé.<br>
Plusieurs méthodes sont possibles.
!!!

````python
tab1 = [4,8,6,0,6,4]
tab2 = [5,1,9,3,7,3]

# le tableau à générer à partir des deux tableaux est [4,5,8,1,6,9,0,3,6,7,4,3]
tab3 = []
max_i = 0
if len(tab1) > len(tab2):
    max_i = len(tab1)
else:
    max_i = len(tab2)
    
for i in range(max_i):
    if i < (len(tab1)):
        tab3.append(tab1[i])
    if i < (len(tab2)):
        tab3.append(tab2[i])

print(tab3)
````

!!! abstract Tableaux à deux dimensions :
Les tableaux en Python peuvent contenir des éléments de n'importe quel type (entiers, flottants, chaînes de caractères, etc...), et même des tableaux. On peut ainsi construire des tableaux de tableaux, comme dans l'exemple ci-dessous, où l'on considère un tableau de 3 tableaux contenant chacun 4 éléments :
!!!

````python
table = [ [1,2,3,4],[5,6,7,8],[9,8,7,6] ]
print(table[1])      # affichage du 2ème tableau
print(table)         # affichage du tableau de tableaux
print(table[2][1])   # affichage du 2ème élément du 3ème tableau
print(len(table))    # affichage du nombre de tableaux du tableau
````

!!! abstract Matrice :
Les tableaux de tableaux sont utiles pour modéliser des tableaux à double entrée (ou des grilles ou des matrices). Pour mieux visualiser les choses, il est préférable d'afficher chacun des tableaux sur une nouvelle ligne comme ci-dessous :
!!!

````python
for tab in table :
    print(tab)
````

Ainsi, chaque tableau correspond à une ligne de la matrice.<br>
Lorsqu'on écrit <code>table[i][j]</code>, <code>i</code> correspond à l'indice de la ligne et <code>j</code> à l'indice de la colonne.

Pour parcourir un tableau à deux dimensions (un tableau de tableaux), il est nécessaire d'utiliser deux boucles <strong>imbriquées</strong> : la boucle externe parcourt toutes les lignes tandis que la boucle interne parcourt toutes les colonnes du tableau.<br>
Le programme ci-dessous parcourt tous les éléments du tableau précédent pour calculer leur somme :

````python
somme = 0
for ligne in table :
    for e in ligne :
        somme = somme + e
print(somme)
````

Le programme ci-dessous montre comment générer un tableau à deux dimensions, ici la table de multiplication des entiers de 0 à 9 :

````python
table_multiplication = [ [0] * 10 for i in range(10)]  #Crée une table de taille 10x10 contenant des 0
for i in range(10):
    for j in range(10):
        table_multiplication[i][j] = i*j

for ligne in table_multiplication:
        print(ligne)
````

Le résultat de chaque multiplication est maintenant stocké dans la table. On peut facilement accéder au résultat de "7x8" comme ci-dessous :

````python
table_multiplication[7][8]
````

Les tableaux à deux dimensions sont souvent utilisés pour représenter des images (tableaux de pixels).
