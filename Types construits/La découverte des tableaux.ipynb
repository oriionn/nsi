<h2 style="text-decoration:underline;" id="partie1";>1- Introduction</h2>

Imaginons que vous souhaitiez écrire un programme en Python qui calcule automatiquement la moyenne de vos notes.
Vous avez par exemple obtenu les notes : 12, 17, 15 et 14.
Vous pourriez procéder de la façon suivante :

```python
a = 12       # on stocke chaque note dans une variable
b = 17
c = 15
d = 14
total = a + b + c + d
nombre_de_notes = 4
moyenne = total / nombre_de_notes
print(moyenne)
```

Cette façon de procéder présente trois défauts qui vont se révéler pénibles si l'on doit rajouter de nouvelles notes :
<ol>
    <li>
        On doit inventer un nouveau nom de variable à chaque fois que l'on ajoute une nouvelle note.
    </li>
    <li>
        Il faut modifier la formule qui calcule le total des notes en ajoutant un terme correspondant à la nouvelle note.
    </li>
    <li>
        Il ne faut pas oublier de modifier aussi le nombre de notes (l'incrémenter de 1).
</ol>

De plus, si l'on souhaite par exemple déterminer de manière automatisée la plus grande des notes, on devra écrire un programme compliqué du type :

```python
maxi = a
if b > maxi:
    maxi = b
elif c > maxi:
    maxi = c
elif d > maxi:
    maxi = d
print(maxi)
```

et il faudra rajouter encore une condition supplémentaire à chaque fois que l'on ajoute une note.
Imaginez la taille du programme si l'on a 100 notes!

Le premier problème (celui des noms de variables) pourrait être résolu en nommant les notes a0, a1, a2, a3, etc... Cette solution ressemble à celle qui a été inventée pour résoudre tous les problèmes à la fois : les **tableaux**. On va ici nommer nos variables : a[0], a[1], a[2], a[3], ... Le nombre entre crochets s'appelle l'<strong>indice</strong>, et pourra être stocké dans une variable.

<h2 style="text-decoration:underline;" id="partie2">2- Les tableaux</h2>

<div style="border:solid;padding:5px;background-color:rgb(255,137,120);color:blue;width:70%">
    <strong>Définition</strong><br>
Un <strong>tableau</strong> permet de stocker plusieurs valeurs dans une seule variable. <br>
On accède aux valeurs individuelles via un <strong>indice</strong> (index en anglais).<br>
La première valeur du tableau correspond à l'indice 0.
</div>

<span style="text-decoration:underline;">Remarques</span>:<br>
<ul>
    <li>
        Dans le langage Python, les <strong>tableaux</strong> sont <strong>dynamiques</strong> : leur taille est variable. Contrairement aux tableaux dans d'autres langages de programmation.
    </li>
    <li>
        Un tableau peut contenir des valeurs de n'importe quel type : on peut considérer des tableaux d'entiers, des tableaux de flottants (<code>float</code>), des tableaux de chaînes de caractères, et même des tableaux de tableaux!<br>
        Dans ce notebook, la plupart des tableaux seront des tableaux d'entiers.
    </li>
    <li>
        Les éléments d'un tableau peuvent être de type différent, même si cette possibilité est assez rarement utilisée en pratique.
    <li>
        Un tableau peut contenir plusieurs fois la même valeur.
    </li>
</ul>

!!! danger Confusion
Le `type` d'un tableau en Python est appelé `list` d'où une confusion fréquente : on retrouve très (trop) souvent le terme de **liste** pour désigner un tableau. Une **liste** est une structure de donnée différente de celle du tableau (elle sera évoquée en terminale).
!!!

Ci-dessous, on stocke 4 valeurs entières dans le tableau <code>a</code> :

```python
a = [12, 17, 15, 14]

print(a)
```

```python
type(a)
```

!!! info Taille d'un tableau
La **taille** du tableau (le nombre de valeurs qu'il contient) s'obtient avec la fonction <code>len</code> (abréviation de "length" qui veut dire "longueur" en anglais).
!!!

```python
len(a)
```

<h3>Retour sur le problème du calcul de la moyenne</h3>

On peut maintenant résoudre le problème du calcul de la moyenne de façon beaucoup plus élégante :

```python
# Vous pouvez rajouter une ou plusieurs notes au tableau de notes
notes = [12,17,15,14]
somme = 0   #somme représente la somme de toutes les notes
nb_notes = len(notes)

for i in range(len(notes)):
    somme = somme + notes[i]
    
moyenne = somme / nb_notes
print(moyenne)
```

Et voilà! Grâce aux tableaux, il est maintenant très simple de rajouter une nouvelle note. Plus besoin de modifier le programme, on ne modifie que le tableau des notes. Essayez!
