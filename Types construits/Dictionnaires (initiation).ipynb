!!! question Question 6. 
Compléter la fonction suivante grâce à sa documentation.
!!!

````python
def nombre_animaux(ferme : dict) -> int :
    '''
    ferme est un dictionnaire dont les clés sont les types d'animaux de la ferme
    et les valeurs associées sont les nombres d'animaux correspondants
    Renvoie le nombre total d'animaux de la ferme 
    '''
    nb = 0
    for valeur in ferme.values():
        nb = nb + valeur
    return nb
````

!!! info Vérifications
Les quatre cellules suivantes permettent de tester votre code. Elles doivent toutes renvoyer `True` lors de leur exécution.
!!!

````python
nombre_animaux({}) == 0
````

````python
nombre_animaux({'lapin':5, 'vache':7, 'cochon':2, 'cheval':4}) == 18
````

````python
nombre_animaux({'lapin':2, 'poule':10, 'âne':4, 'mouton':25}) == 41
````

````python
nombre_animaux(ferme_charles) == 36
````

!!! question Question 7. 
Compléter la fonction ci-dessous grâce à sa documentation.
!!!

````python
def nombre_animal(ferme : dict, animal : str) -> int :
    '''
    ferme est un dictionnaire dont les clés sont les types d'animaux de la ferme
    et les valeurs associées sont les nombres d'animaux correspondants
    animal est une chaîne de caractères donnant le nom d'un animal
    Si animal est dans ferme, renvoie le nombre d'animaux correspondant ; renvoie 0 sinon
    '''
    if animal in ferme:
        return ferme[animal]
    else:
        return 0
````

!!! info Vérifications
Les quatre cellules suivantes permettent de tester votre code. Elles doivent toutes renvoyer `True` lors de leur exécution.
!!!

````python
nombre_animal({}, 'vache') == 0
````

````python
nombre_animal({'lapin':5, 'vache':7, 'cochon':2, 'cheval':4}, 'lapin') == 5
````

````python
nombre_animal(ferme_charles, 'cheval') == 4
````

````python
nombre_animal(ferme_charles, 'canard') == 0
````

<h2 class="alert alert-info">Exercice 4 : jours et mois</h2>

!!! info Mise ne situation
On se donne ci-dessous deux tableaux `jour` et `mois` de douze éléments :
 - le tableau `jour` contient le nombre de jours de chaque mois de l'année ;
 - le tableau `mois` contient les noms des douze mois de l'année.

Exécuter la cellule suivante :
!!!

````python
jour = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
mois = ['Janvier', 'Février', 'Mars', 'Avril', 'Mai', 'Juin', 'Juillet', 'Août', 'Septembre', 'Octobre', 'Novembre', 'Décembre']
````

!!! question Question :
Ecrire dans la cellule ci-dessous un programme permettant de définir le dictionnaire suivant :
```python
{'Janvier': 31, 'Février': 28, 'Mars': 31, 'Avril': 30, 'Mai': 31, 'Juin': 30, 'Juillet': 31, 'Août': 31, 'Septembre': 30, 'Octobre': 31, 'Novembre': 30, 'Décembre': 31}
```
!!!

````python
# Ecrire l'instruction ici
dico = {}
for i in range(len(mois)):
    key = mois[i]
    value = jour[i]
    dico[key] = value
print(dico)
````

<h2 class="alert alert-info">Exercice 5 : traduction</h2>

!!! info Exécuter la cellule suivante.
!!!

````python
dictionnaire_anglais_français = {'computer':'ordinateur', 'mouse':'souris', 'screen':'écran'}
````

!!! question Question :
Proposer un programme ci-dessous afin que la variable `dictionnaire_français_anglais` contienne le dictionnaire
```python
{'ordinateur': 'computer', 'souris': 'mouse', 'écran': 'screen'}
```
!!!

````python
#le programme ici
dictionnaire_français_anglais = {}
for anglais, français in dictionnaire_anglais_français.items():
    dictionnaire_français_anglais[français] = anglais
    
print(dictionnaire_français_anglais)
````
