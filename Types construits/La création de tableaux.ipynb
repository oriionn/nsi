!!! abstract a) En écrivant la liste complète des éléments

!!!

```python
mon_tableau = [4,10,-9,7,5.4]
print(mon_tableau)
print(mon_tableau[2])
print(len(mon_tableau))
```

!!! abstract 
Parfois, on ne connait pas à l'avance les éléments du tableau. Cela peut se produire par exemple lorsqu'on demande à un utilisateur de saisir des valeurs que l'on souhaite stocker en mémoire.
On peut alors procéder de la façon suivante :<br>
    <ol>
    <li>
        Initialiser un tableau ayant la taille souhaitée, avec par exemple la valeur 0
    </li>
    <li>
        Ecrire l'un après l'autre les éléments du tableau
    </li>
    </ol>
<strong>Exemple :</strong>
!!!

```python
valeurs = [0] * 5    # Ceci crée un tableau de 5 éléments ayant tous la valeur 0
print("Le tableau initial valeurs : ",valeurs)
for i in range(5):
    valeurs[i] = int(input("Saisissez un nombre entier puis tapez sur Entrée : "))
print("Voici le tableau avec les nombres que vous avez saisis :",valeurs)
```

!!! abstract b) En ajoutant un à un les éléments du tableau
Lorsqu'on ne connait pas à l'avance le nombre d'éléments du tableau, la méthode précédente ne fonctionne pas. On peut alors créer un tableau vide, puis la rallonger petit à petit en lui ajoutant des éléments les uns après les autres, au moyen de la fonction <code>append</code> ("append" signifie "ajouter, joindre" en anglais) : 
!!!

```python
mon_tableau = []
mon_tableau.append(6)
mon_tableau.append(7)
mon_tableau.append(10)
print(mon_tableau)
print(mon_tableau[2])
print(len(mon_tableau))
mon_tableau.append(4)
print(mon_tableau)
```

Le programme ci-dessous demande à l'utilisateur de saisir des valeurs jusqu'à ce qu'il saisisse la valeur 0 :

```python
n = 1
mon_tableau = []
while n != 0:
    n = int(input("Saisissez un nombre entier (0 pour arrêter) : "))
    if n != 0:
        mon_tableau.append(n)
print("Voici le tableau des nombres saisis :",mon_tableau)    
```

!!! quote Remarque : 
Il existe aussi en Python :
* une fonction d'insertion qui permet d'insérer un élément à n'importe quelle position dans le tableau
* des fonctions de suppression qui permettent de supprimer n'importe quel élément du tableau. 
Les éléments consécutifs d'un tableau sont stockés dans des emplacements contigus de la mémoire de la machine. Les fonctions d'insertion et de suppression obligent donc la machine à décaler d'un cran vers la gauche tous les éléments situés à droite de l'emplacement d'insertion ou de suppression. Il faut donc éviter d'utiliser ces fonctions trop fréquemment ou sur des tableaux très grands. Si l'on doit faire de nombreuses insertions et/ou suppressions, on utilisera une structure plus sophistiquée qu'un tableau, appelée une liste chaînée (hors programme de 1ère Spécialité NSI).<br>

Voici quelques exemples d'utilisation des fonction d'insertion /suppression :
!!!

```python
tab = [2,4,6,8,4,10] # pop, del et remove
print("insère la valeur 5 à l'indice 3, donc entre les valeurs 6 et 8")
tab.insert(3,5)   # insère la valeur 5 à l'indice 3, donc entre les valeurs 6 et 8
print(tab)
print("supprime la première valeur 4 de tab")
tab.remove(4)     # supprime la première valeur 4 de tab
print(tab)
print("supprime la valeur d'indice 3 de tab  ")
del tab[3]        # supprime la valeur d'indice 3 de tab  
print(tab)
print("extraction d'un élément à l'indice 2")
element = tab.pop(2)
print("élément extrait : ", element)
print(tab)
```

!!! abstract Fusion de tableaux : 
<br>On peut fusionner ou concaténer deux tableaux en les séparant par l'opérateur + . Cela renvoie un nouveau tableau qui est la concaténation des deux tableaux.<br>
Voir l'exemple ci-dessous :
!!!

```python
tab1 = [2,4,6,8]
tab2 = [7,8,9]
gros_tab = tab1 + tab2
print(gros_tab)
gros_tab = tab2 + tab1     # l'addition des tabs n'est pas commutative!
print(gros_tab)            # les tableaux tab1 et tab2 n'ont pas été modifiées  
```

On peut aussi étendre un tableau en lui adjoignant un autre tableau à l'aide de la fonction <code>extend</code> comme l'illustre l'exemple ci-dessous :

```python
tab1 = [2,4,6,8]
tab2 = [7,8,9]
tab1.extend(tab2)   # on ne crée pas de nouveau tableau, mais tab1 est modifié
print(tab1)
print(tab2)   #tab2 n'est pas modifié
```

<h3 style="text-decoration:underline;">c) En utilisant la fonction <code>range</code></h3>

La fonction <code>range</code> (souvent utilisée dans les boucles <code>for</code>) génère automatiquement un tableau d'entiers. <br><br>
Elle peut prendre un, deux ou trois paramètres entiers :<br>
<ul>
    <li>
        <code>range(n)</code> renvoie tous les entiers compris entre 0 inclus et n non inclus.
    </li>
    <li>
        <code>range(p,n)</code> renvoie tous les entiers compris entre p inclus et n non inclus.
    <li>
        <code>range(p,n,pas)</code> renvoie tous les entiers compris entre p inclus et n non inclus avec un pas.<br>
        Le pas peut être négatif. Dans ce cas, p doit être plus grand que n car p est la valeur de départ.
    </li>
 </ul>
        
Remarque : en réalité, la fonction <code>range</code> renvoie un objet qui n'est pas tout à fait un tableau. Pour obtenir un vrai tableau, il faut convertir cet objet en tableau avec la fonction <code>list</code>.

```python
tab1 = list(range(8))
print(tab1)
tab2 = list(range(10,20))
print(tab2)
tab3 = list(range(5,25,3))
print(tab3)
tab3 = list(range(25,3,-3))
print(tab3)
```

<h3 style="text-decoration:underline;">d) Génération d'un tableau en compréhension</h3>

En mathématiques, on utilise souvent des formulations du type <strong>"l'ensemble de tous les nombres de la forme 4n+3, pour n entier compris entre 0 et 5"</strong>.
Cela correspond à l'ensemble de nombres suivant : { 3 , 7 , 11 , 15 , 19 , 23 } <br>
Le langage Python permet de définir un tableau selon le même principe, avec une syntaxe très courte. Ainsi, le tableau des nombres précédents peut être générée <strong>en compréhension</strong> de la façon suivante :

```python
mon_tab = [ 4*n+3 for n in range(6)]
print(mon_tab)
```

!!! question Exercice 5 :

!!!

<h3 style="text-decoration:underline;">e) Filtrage d'un tableau</h3>

On peut créer un tableau à partir d'un tableau déjà existant en ne sélectionnant que certains éléments du tableau d'origine qui vérifient une certaine condition.

<strong>Sélection avec une condition portant sur les indices : le slice</strong>

"slice" signifie "couper en tranches" en anglais. On ne sélectionne qu'une portion du tableau comprise entre deux indices. 
<code>mon_tableau[a:b]</code> sélectionne la tranche qui va de l'indice `a` inclus à l'indice `b` exclu.<br>
On peut ajouter un pas :
<code>mon_tableau[a:b:c]</code> ici `c` correspond au pas<br>
Si le pas est négatif, ltableau est renversé.

```python
mon_tableau = list(range(25))
print(mon_tableau)
tab1 = mon_tableau[3:6]   # sélection des élements de l'indice 3 inclus à l'indice 6 exclu
print(tab1)
tab2 = mon_tableau[:10]   # sélection des 10 premiers éléments (de l'indice 0 à l'indice 9)
print(tab2)
tab3 = mon_tableau[15:]   # sélection de tous les éléments à partir de celui d'indice 15
print(tab3)
tab4 = mon_tableau[::2]   # sélection de tous les éléments d'indice pair (de 2 en 2)
print(tab4)
tab5 = mon_tableau[1::2]  # sélection de tous les éléments d'indice impair (de 2 en 2 en commençant à 1)
print(tab5)
tab6 = mon_tableau[::-1]  # tableau obtenu en parcourant tous les éléments à partir de la fin
print(tab6)
tab7 = mon_tableau[::-2]  # tableau obtenu en parcourant tous les éléments à partir de la fin de 2 en 2
print(tab7)
```

<strong>Sélection avec une condition portant sur les valeurs</strong>

Lors de la génération d'un tableau en compréhension, on peut ajouter une condition (avec le mot-clé <code>if</code>) qui permet d'exclure certaines valeurs du tableau.<br>
Par exemple, le programme ci-dessous ne retient que les valeurs supérieures à 10 du tableau initial :

```python
tableau_initial = [15,12,7,9,10,18,6]
nouveau_tableau = [ n for n in tableau_initial if n >= 10 ]
print(nouveau_tableau)
```
