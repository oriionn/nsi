!!! info Turtle
Python propose un certain nombre d'instructions primitives. Le langage contient également de très nombreuses bibliothèques (math, random, etc). Dans ce notebook nous allons (re)découvrir la bibliothèque TURTLE. Ce clone du langage éducatif LOGO permet de déplacer une tortue munie d'un crayon sur la surface d'une feuille virtuelle.
Les instructions de bases seront :

| instruction  | Description                                   |
| ------------ | --------------------------------------------- |
| goto(x, y)   | aller au point de coordonnées (x, y)          |
| forward(d)   | avancer de la distance d                      |
| backward(d)  | reculer de la distance d                      |
| left(a)      | pivoter à gauche d'un angle a                 |
| right(a)     | pivoter à droite d'un angle a                 |
| circle(r, a) | tracer une arc de cercle de rayon r d'angle a |
| dot(r)       | tracer un point de rayon r                    |
!!!

```python
# exemple
from turtle import *
forward(60)
left(120)
forward(60)
right(90)
circle(60, 300)
right(90)
forward(60)
goto(0, 0)
ht() #hide turtle (optionnel)
done()   #instruction nécessaire dans un notebook uniquement
```

S'ajoute à cette base une série d'instructions avancées (voir documentation : https://docs.python.org/fr/3/library/turtle.html) :

| instruction  | Description                                   |
| ------------ | --------------------------------------------- |
| up()         | relever le crayon                             |
| down()       | redescendre le crayon                         |
| width(e)     | fixer l'épaisseur e du crayon                 |
| color(c)     | sélectionner la couleur c du trait            |
| begin_fill() | activer le mode remplissage                   |
| end_fill()   | désactiver le mode remplissage                |
| fill_color() | sélectionner la couleur de remplissage        |

```python
# exemple :
#rectangle gris épais
width(6)
color("grey")
goto(60, 0)
goto(60, 110)
goto(0, 110)
goto(0, 0)
#déplacement
up()
goto(5, 5)
down()
#sablier gris clair
width(1)
fillcolor("lightgrey")
begin_fill()
goto(55, 5)
goto(5, 105)
goto(55, 105)
goto(5, 5)
end_fill()
ht()   #hide turtle
done()
```
