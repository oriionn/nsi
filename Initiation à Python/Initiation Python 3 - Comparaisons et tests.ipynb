!!! info Faire des choix

Dans une suite d'instructions on est souvent amené à faire des choix.

**Si** une ou plusieurs conditions sont réalisées **Alors** on exécute une ou plusieurs instructions


La structure en Python: On indente (4 espaces) les instructions qui doivent s'exécuter si la condition est vérifiée

```py
if condition(s) :
    instruction1
    instruction2

```

**Voici un exemple qui affiche si les variables a et b  sont des entiers pairs**
!!!

````python
a = 12
if a % 2 == 0:
    print(a, " est un entier pair")
    
b = 13
if b % 2 == 0:
    print(b," est un entier pair")
````

Il est possible de combiner des conditions. Par exemple, on veut vérifier que la saisie d'une moyenne d'un élève ne peut être inférieure à 0 ni supérieure à 20.

````python
note = -5
# première version avec un "OU"
if note < 0 or note > 20:
    print("Cette note n'est pas valable")
    
# une autre possibilité avec "ET":
if note >=0 and note <=20:
    print("Cette note est valable")
````

!!! info Si...Sinon...

```py
    if condition(s):
        instruction1
        instruction2
    else:
        instruction3
```
 
 Dans cette structure, si la ou les conditions sont vraies, les instructions 1 et 2 sont exécutées.
 Sinon c'est l'instruction3 qui est exécutée.
 
 **Voici un exemple**
!!!

````python
a = 12

if a%2 == 0:
    print(a, " est un entier pair")
else:
    print(a," est un entier impair")
````

!!! info Si...SinonSi...SinonSi...Sinon

```py
    if condition1:
        instruction1
    elif condition2:
        instruction2
        instruction3
    elif condition3:
        instruction4
    else:
        instruction5
```

Dans cette structure on teste plusieurs conditions
!!!

!!! question Exercice 1 : compter les points au mölky

Au jeu du mölky, chaque joueur marque à son tour de jeu entre 0 et 12 points, qui viennent s'ajouter à son score précédent. Le premier à atteindre un score de 51 gagne. Mais gare ! Quiconque dépasse le score cible de 51 revient immédiatement à 25 points.

Compléter le programme demandant le score actuel et le nombre de points marqués, et qui affiche le nouveau score ou signale une éventuelle victoire.

!!!

````python
score = int(input("Entrer le score actuel : "))
points = int(input("Entrer le nombre de points marqués : "))

if points > 12: 
    print("C'est pas bien de tricher !!!!")
else:
    new_score = score + points

    if new_score > 51:
        new_score  = 25

    print("Nouveau score : " + str(new_score))
````
