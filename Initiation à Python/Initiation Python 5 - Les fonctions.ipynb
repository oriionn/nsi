!!! info Une fonction permet de délocaliser une partie de code et éventuellement de la réutiliser

**Syntaxe**

En général, une fonction reçoit des données en paramètres (mais ce n'est pas toujours le cas) en renvoie TOUJOURS des données<br>

```py
def ma_fonction(parametre1,parametre2,...):
    instructions
    return resultat
```
Pour utiliser une fonction, il faut l'appeler.

```py
print(ma_fonction(a,b,...))
# affiche le résultat de la fonction
```

**Exemple**
!!!

````python
# fonction qui renvoie le plus grand de 2 nombres entiers passés en paramètres
def plus_grand2(a ,b ):
    if a > b:
        return a
    else:
        return b
    
print(plus_grand2(7,12))
````

!!! info Bonnes pratiques
Lorsque l'on rédige une fonction nous veillerons à avoir de bonnes pratiques :
 * Le nom des fonctions sera explicite et rédigé en `snake case` : mots en minuscule et séparé par un `_`
 * Dès que c'est possible, nous indiquerons le type de données attendues en paramètre et/ou en sortie de la fonction 
 ```python
def plus_grand(a : int, b : int) -> int :

```
Ici nous venons d'indiqué que les paramètres `a` et `b` seront des entiers et que la fonction renvoie un entier également
* Nous mettrons toujours une courte documentation de la fonction (docstring) entre des triples guillements
 ```python
def plus_grand(a : int, b : int) -> int :
    """une fonction qui compare deux entiers et renvoie le plus grand """

```
* La fonction précédente sera correctement rédigée ainsi :
```python
def plus_grand(a : int, b : int) -> int :
    """une fonction qui compare deux entiers et renvoie le plus grand """
    if a > b:
        return a
    else:
        return b
```

!!! question Exercices niveau 1

* Écrire une fonction `plus_petit` qui prend en paramètres deux nombres réels et qui renvoie le plus petit.
* Écrire une fonction `plus_grand3` qui prend en paramètres trois nombres entiers et qui renvoie le plus grand.
* Écrire une fonction `somme3` qui prend en paramètres trois nombres réels et qui renvoie leur somme.

!!!

````python
# code ici
def plus_petit(a: int, b: int) -> int:
    """Retourne le nombre entier le plus petit entre a et b"""
    if a < b:
        return a
    else:
        return b
````

````python
# code ici
def plus_grand3(a: int, b: int, c: int) -> int:
    """Retourne le nombre le plus grand entre a, b et c"""
    list = [a,b,c]
    list.sort()
    return list[2]
````

````python
# code ici
def somme3(a,b,c):
    """Retourne la somme de trois nombres entiers: a, b et c"""
    return a + b + c
````

!!! question Exercices niveau 2

* Écrire une fonction `produit_n` qui prend en paramètre un entier n et qui renvoie $1 \times2 \times3\times...\times n$.
* Écrire une fonction `est_pair` qui prend en paramètre un entier n et qui renvoie True si il est pair et False sinon.
* Écrire une fonction `table_multiplication` qui prend en paramètre un entier n et qui affiche la table de multiplication de n jusqu'à 10. (on utilise print à la place de return)
!!!

````python
# code ici
def produit_n(n: int) -> int:
    """Renvoie 1x2x3... xn"""
    value = 1
    for i in range(1, n):
        value = value * (i + 1)
    return value
````

````python
# code ici
def est_pair(n: int) -> bool:
    """Retourne si n est pair ou non"""
    return n % 2 == 0
````

````python
# code ici
def table_multiplication(n: int):
    """Affiche la table multiplication de n"""
    for i in range(11):
        print(i, "x", n, "=", i * n)
````

!!! question Exercice niveau 3

Écrire une fonction `triangle(n)` traçant avec Turtle un triangle équilatéral noir tel le suivant, avec un coté de longueur `n`.

!!!

````python
# code ici
from turtle import forward, done, left, right, begin_fill, end_fill

def triangle(n: int):
    """Produit un triangle équilatérale de coté n"""
    begin_fill()
    forward(n)
    for i in range(2):
        left(120)
        forward(n)
    end_fill()
````
